```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('books').getOrCreate()
spark
```

```python
from pyspark.sql import functions as func
```

```python
inputDF=spark.read.text("book.txt")
inputDF
```

```python
inputDF.show(5)
```

```python
inputDF.value
```

```python
words=inputDF.select(func.explode(func.split(inputDF.value,"\\W+")).alias("word"))
words.printSchema()
```

```python
words.show(250)
```

```python
words=words.filter(words.word!="")
words.show(250)
```

```python
lowercase=words.select(func.lower(words.word).alias("word"))
```

```python
wordCounts=lowercase.groupBy("word").count().sort(func.desc("count"))
wordCounts.show(500)
```
